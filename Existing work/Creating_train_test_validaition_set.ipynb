```python
def train_val_test_set():
    tot_kill = pd.DataFrame()
    tot_death = pd.DataFrame()

    for i in range(1,501):
        tot_kill = tot_kill.append({'videos': 'frame_kill_'+str(i)+'.jpg'}, ignore_index=True)
        tot_death = tot_death.append({'videos': 'frame_death_'+str(i)+'.jpg'}, ignore_index=True)

    tot_kill['label'] = 1
    tot_death['label'] = 0
    
    train_set_kill = tot_kill[:300]
    train_set_death = tot_death[:300]
    train_set = train_set_kill.append(train_set_death, ignore_index=True)
    
    val_set_kill = tot_kill.iloc[300:400]
    val_set_death = tot_death.iloc[300:400]
    val_set = val_set_kill.append(val_set_death, ignore_index=True)
    
    test_set_kill = tot_kill.iloc[400:500]
    test_set_death = tot_death.iloc[400:500]
    test_set = test_set_kill.append(test_set_death, ignore_index=True)
    
    return train_set, val_set, test_set
```

```python
train_set, val_set, test_set = train_val_test_set()
```
